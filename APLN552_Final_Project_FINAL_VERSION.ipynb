```python
from google.colab import files
from transformers import pipeline
import pandas as pd
pd.set_option('display.max_colwidth', None)
```

```python
uploaded = files.upload()
```

```python
import io

df = pd.read_excel(io.BytesIO(uploaded['DRINKS.xlsx']))
table = df.astype(str)
```

```python
tqa = pipeline(task = 'table-question-answering',
                  model = 'google/tapas-base-finetuned-wtq')
```

```python
query = "Show me one bright gin cocktail"
print(tqa(table=table, query=query)['cells'][0])
drink = str(tqa(table=table, query=query)['cells'][0])
find_drink = table.loc[table['DRINK_NAME'] == drink]
display(find_drink['INSTRUCTIONS'])
```
